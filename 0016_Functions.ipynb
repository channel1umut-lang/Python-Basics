```python
#args, #kwargs, #arguments, #key word arguments
```

```python
def args_sum(*args):# sayisi belli olmayan bir unoutum varsa args kullanabilirz.
  return sum(args)
```

```python
args_sum(10,20,30,40,50,60)
```

```python
def args_example(*args):
  print(args)
```

```python
args_example(10,20,1412,404,44)
```

```python
def kwargs_example(**kwargs):
  print(kwargs)
```

```python
kwargs_example(apple=100,banana=150,melon=200)
```

```python
def kwargs_example_2(**kwargs):
  if "apple" in kwargs:
    print("appleeee")
  else:
    print(":(")
```

```python
kwargs_example_2(apple=10, banana = 20)
```

```python
kwargs_example_2( banana = 20)
```
