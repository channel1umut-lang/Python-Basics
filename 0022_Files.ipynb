```python
# Yazmak istediğiniz metni tanımlayın
metin = "Text 1 \nText 2 \nText 3 \n"

# Dosyayı yazma modunda açın
with open('ornek_dosya.txt', 'w') as dosya:
    dosya.write(metin)
with open('ornek_dosya.txt', 'r') as dosya:
    icerik = dosya.read()

print("Dosya oluşturuldu!")
```
