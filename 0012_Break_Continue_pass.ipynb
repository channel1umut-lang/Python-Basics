```python
#continue - break - pass
```

```python
my_list = [10,20,30,40,50,60,70]
```

```python
print("loop started")
for num in my_list:
  print(num)
print ("loop ended")
```

```python
40 in my_list
```

```python
for num in my_list:  # donguyu kirdi
  print(num)
  if num == 40:
    print("yes")
    break
```

```python
my_list2 = [10,20,30,40,50,60,70]
```

```python
for num2 in my_list2: # continue o anda num2 40 a esitse esit oldugu durumdaki kodu calistirmiyor. Donguye devam ediyor.
  print(num2)
  if num2 == 40:
    continue
    print("yes")
```

```python
for num2 in my_list2: # continue o anda num2 40 a esitse esit oldugu durumdaki kodu calistirmiyor. Donguye devam ediyor.
  if num2 == 40:
    continue
  print(num2)
  print("yes")
```
