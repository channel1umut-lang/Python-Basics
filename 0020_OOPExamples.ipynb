#inheritance

```python
class Musician():
  def __init__(self,name):
    self.name=name
    print("musician class")
  def test1(self):
    print("test1")
  def test2(self):
    print("test2")
```

```python
atil = Musician("Atil")
```

```python
atil.name
```

```python
atil.test1()
```

```python
class MusicianPlus(Musician): #diger sinifin butun ozelliklerini buraya aktardik. atil. yazdigimizda yanda cikan hersey atlas. yazinca da otomatik geliyor.
  def __init__(self,name):
    Musician.__init__(self,name)
    print("musician plus")
  def test3(self):
    print("test3")
  def test1(self):
    print("test1 test1 test1") # overwrite
```

```python
atlas = MusicianPlus("Atlas")
```

```python
type(atlas)
```

```python
atlas.name #initiate oldugu icin parantez acmiyoruz
```

```python
atlas.test1()
```

```python
atlas.test2()
```

```python
atlas.test3() #atlas. yaptigimda test3 de eklenmis halde cikti ama atil. yaptigimda test3 ona dahil edilmedi.
```

```python
atil.test1()
```

#Polymorphism

```python
class Banana():
  def __init__(self,name):
    self.name=name
  def info(self):
    return f"100 calories {self.name}"
class Apple():
  def __init__(self,name):
    self.name=name
  def info(self):
    return f"150 calories {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruit_list = [banana,apple]
```

```python
for fruit in fruit_list: #fruit degiskeni .infoyu calistirabildi
  print(fruit.info())
```

#encapsulation

```python
class Phone():
  def __init__(self,name,price):
    self.name=name
    self.__price=price #private yaptik fiyati asla degismiyor
  def info(self):
      print(f"{self.name} price is: {self.__price}")
  def change_price(self,price): #degistirilebilir bir kod istiyorsak bunu yazabiliriz.
    self.__price=price
```

```python
iphone = Phone("iPhone 14",500)
```

```python
iphone.info()
```

```python
iphone.price = 4 #fiyati degismiyor.
```

```python
iphone.info()
```

```python
iphone.change_price(400)
```

```python
iphone.info()
```

#abstraction

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
  @abstractmethod
  def max_speed(self):
    pass
```

```python
#my_car = Car() bunu oolusturamiyorum cunku abstract icine atamam
```

```python
class Tesla(Car): # Car sinifini kullandigimiz icin altindaki def max speed i kullanmak zorundayim
  def max_speed(self):
    print("200km")
```

```python
tesla = Tesla()
```

```python
tesla.max_speed()
```

```python
class Mercedes(Car):
  def max_speed(self):
    print("250km")
```

```python
merce = Mercedes() # bu kod calismadi cunku max_speed i calistirmak zorundayiz cunku olusturdugumuz Car clasi abstract yapida bir sinif.
```

```python
class fruit():
  def __init__(self,name,calories):  # amac pythonun kendi icindeki tanimli fonksiyonlari degistirmek icin
    self.name=name
    self.calories=calories
  def __str__(self):
    return f"{self.name}: {self.calories} calories"
  def __len__(self):
    return self.calories
```

```python
class Train():
  def __init__(self,name):
    self.name=name
  def __getitem__(self,key):
    if key=="a":
      return self.name
    else:
      return "not found"
```

```python
myTrain = Train("myTrain")
```

```python
myTrain["a"]
```
