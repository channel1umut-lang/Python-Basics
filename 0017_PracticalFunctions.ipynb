```python
def divideNumber(number):
  return number/2
```

```python
divideNumber(20)
```

```python
myList = [3,5,7,10,20,30]
```

```python
my_result_list = []
for i in myList:
  my_result_list.append(divideNumber(i))
```

```python
def myfunc(a, b):
  return a + b
```

```python
x = map(myfunc, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple'))
print(x)
```

```python
#convert the map into a list, for readability:
print(list(x))
```

```python
map(divideNumber,myList)
```

```python
print(list(map(divideNumber,myList)))
```

```python
def control_string(string):
  return "atil" in string #stringin icinde atil geciyor mu bunu kontrol ediyoruz
```

```python
control_string("atil")
```

```python
control_string("samancioglu")
```

```python
myStringList = ["atil","atil samancioglu","samancioglu","atlas"]
```

```python
list(map(control_string,myStringList))
```

#FILTER

```python
list(filter(control_string, myStringList)) # sadece atil gecenleri gosterdi.
```
