```python
try:
  myAge = int(input("enter age : "))
  print(myAge * 2)
except: # sadece except yazdigimizda herhangi bir hatayi kabul ediyor. ValueError: yazdigimizda sadece value error alirsa kabul ediyor
  print("Please enter a number")
```

```python
while True:
  try:
     myAge = int(input("enter age : "))
     print(myAge * 2)
     break# donguyu durdurduk sayi girerek
  except:
     print("Please enter a number")
  else:
    print("else exacuted")
    #break
```
