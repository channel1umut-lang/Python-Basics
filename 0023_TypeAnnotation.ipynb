```python
def process_value(value: int | str) -> str:
  if isinstance(value, int):
    return f"processed integer value: {value}"
  elif isinstance(value, float):
    return f"processed float value: {value}"
  else:
    return f"processed string value: {value}"
```

```python
process_value(4)
```

```python
process_value("umut")
```

```python
process_value(4.14)
```
