```python
class Person():
  #Property
  #name = ""
  #age  = 0
  #gender = ""
  job = "developer" #bu siniftaki herkesi developer olarak alabiliriz istesek
  # zorunlu olmayan ama istefdigimizde degistirebilecegimiz parametreleri buraya yaziyoruz
  #initializer methodu
  def __init__(self,name,age,gender): #ozel fonksiyonlar__ seklinde tanimlanir. self sinifa referans veren keyword sadece karisiklik olmamasi icin
    self.name = name #zorunlu olmasi gereken parametreleri buraya yaziyoruz
    self.age = age
    self.gender = gender
    print("init method has been executed")
  #method
  def test(self):
    print("test method has been executed")
  def printName(self):
    print(self.name)
```

```python
atil = Person("umut",29,"male")
```
