```python
my_list = [20,30,40,50,60]
```

```python
for num in my_list:
  print(num)
```

```python
len(my_list)
```

```python
for ix in range(len(my_list)): # len(my_list) burda 5 e denk geliyor range de 5 e kadar olana sayilari yazdiriyor.
  print(ix)
```

```python
for ix in range(len(my_list)):
  print(my_list[ix])   # ix inci elemani yazdiriyor. ix 0 oldugunda 20 yi yazdiryor.
```

```python
#enumerate
```

```python
for element in enumerate(my_list):
  print(element)
```

```python
for (ix,value) in enumerate(my_list):
  print(value)
```

```python
food_list = ["apple", "banana", "melon"]
```
