```python
#Global Eger hicbirinde yoksa built-ine yani pythona bakiyor
myString = "Atıl"

def myFunction():
    #Enclosing
    myString = "Atıl 2"
    print(myString)

    def myFunction2():
        #Local
        myString = "Atıl 3"
        print(myString)
    myFunction2()
```
